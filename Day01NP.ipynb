#Aggregate functins (min,max,avg,count ect)

```python
#sum
arr=np.array([5,3,7,8,9])
sumarray=np.sum(arr)
print((sumarray))
```

```python
#mean
arr=np.array([10,50,20,30,40])
avgarray=np.mean(arr)
print(avgarray)
```

```python
#median
arr=np.array([10,20,30,40,50,60])
med=np.median(arr)
print(med)
```

```python
arr=np.array([10,20,30,40,50,60])
sd=np.std(arr)
print(sd)
```

```python
arr=np.array([10,20,30,40,50,60])
v=np.var(arr)
print(v)
```

```python
#product
arr=np.array([10,20,30])
product=np.prod(arr)
print(product)
```

```python
#max / min
arr=np.array([10,20,30,40])
print("max",np.max(arr))
print("min",np.min(arr))
```

```python
#arg max arg min it provide value index of max and min value
arr=np.array([10,20,30,40])
print("argmax",np.argmax(arr))
print("argmin",np.argmin(arr))
```
