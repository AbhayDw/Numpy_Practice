**SHUFFLE AND PERMUTATION**

```python
#shuffle it also followed in place terminology
arr=np.array([10,20,30,40,50])
rng.shuffle(arr)
arr
```

```python
#permutation it avoid in place terminology and also perform operations in diff veriable
arr=np.array([10,20,30,40,50])
shuffled_arr=rng.permutation(arr)
print(shuffled_arr)
print(arr)
```

**PERMUTATED/PERMUTATION WHICH IS USED IN 2D ARRAY (axis=0 means ROW and axis=1 means column)**

```python
arr=np.arange(1,26,1).reshape(5,5)
arr
```

```python
#row wise shuffle use axis=0
rng.permutation(arr,axis=0)
```

```python
#column wise shuffle use axis=1
rng.permutation(arr,axis=1)
```
